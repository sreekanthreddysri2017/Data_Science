```python
y_test #actual

# # Visualizing the results
```

```python
plt.scatter(x_train, y_train) #plotting the observation line
plt.plot(x_train, regressor.predict(x_train)) #plotting the regression line
plt.xlabel('Years of experience')
plt.ylabel('Salaries')
```

```python
# # Evaluating the model
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
print('MSE',mean_squared_error(y_test,y_pred))
```

```python
from math import sqrt as sqrt
print('RMSE',sqrt(mean_absolute_error(y_test,y_pred)))
```

```python
r2=r2_score(y_test,y_pred)
print(r2)
```

```python
import pandas as pd
import seaborn as sn
```

```python
df=pd.read_csv('MLR_car_data.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# Data Preprocessing
```

```python
df_required=df.drop('Car_Name', axis=1)
df_required
```

```python
df_required['Present_year']=2022
df_required.head()
```

```python
df_required['Age_year']=df_required['Present_year']-df_required['Year']
df_required.head()
```

```python
df_required.drop(['Year','Present_year'], axis=1, inplace=True)
df_required.head()
```

```python
df_required=pd.get_dummies(df_required)
df_required.head()
```

```python
df_required_x=df_required.drop('Selling_Price', axis=1) #because i want only features for checking correlation
df_required_x.head()
```

```python
# Multi collinearity
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sn.heatmap(df_required_x.corr(), annot=True) #annot-an array of the same shape as data which is used to annotate the heatmap
```

```python
# # Set up the dependant and the independant variables
```

```python
x=df_required_x
y=df_required.iloc[:,0]
```

```python
x
```

```python
y
```

```python
# Divide the data into train and test sets
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
# Train the algorithm
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred=pd.DataFrame(y_pred,columns=['predicted'])
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
r2= r2_score(y_test,y_pred)
print(r2)
```

```python
import pandas as pd
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
features=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
```

```python
x=df[features]
```

```python
y=df.label
```

```python
y
```

```python
# Split x & y into training and testing sets
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=0)
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Evaluating the performance of a model
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(y_test,y_pred)
confusion_matrix = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[False, True])
confusion_matrix.plot()
plt.show()
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#Precision
from sklearn.metrics import precision_score
precision_score(y_test, y_pred)#, average=None)
```

```python
# Recall
from sklearn.metrics import recall_score
recall_score(y_test, y_pred)#, average=None)
```

```python
F1_score = metrics.f1_score(y_test, y_pred)
print(F1_score)
```
